```python
#to crawl till page 10 only

ROOT_URL=f"http://ufcstats.com/statistics/events/completed?page="

for n in range(2,11):
    url_page = f"{ROOT_URL}{n}"
    print(url_page)

    response = requests.get(url_page)
    soup = BeautifulSoup(response.text, "html.parser")

    #go each link in page 
```

```python
#go to page one
#go to match link
#in match link search for lightweight category and method sub or ko or tko only
#go into that matched category
#crawl all fight detail in the table
```

```python
table_rows_match = soup.find_all("tr",{"class":"b-statistics__table-row"})
```

```python
num_row =len(table_rows_match)
```

```python
#go each match link in page for page 1
#url

match_urls= []
for match in range(0,num_row):
    match_link= table_rows_match[match].find("a")
    match_urls.append(match_link.find_all("href"))
    
```
