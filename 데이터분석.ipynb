## 가설7. ko경기의 kd 발생은 nonko 경기의 kd 발생보다 많다. -> 기각

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt

# 1) KD binary 생성 (KD 있으면 1, 없으면 0)
ko_data['kd_binary']  = (ko_data['winner_kd'] > 0).astype(int)
nko_data['kd_binary'] = (nko_data['winner_kd'] > 0).astype(int)

# 2) 값 집계
ko_kd_count  = ko_data['kd_binary'].sum()
ko_no_kd     = len(ko_data) - ko_kd_count

nko_kd_count = nko_data['kd_binary'].sum()
nko_no_kd    = len(nko_data) - nko_kd_count

# 3) 카이제곱 검정용 테이블 만들기
contingency_table = [
    [ko_kd_count,  ko_no_kd],     # KO 경기: KD 있음 vs 없음
    [nko_kd_count, nko_no_kd]     # NKO 경기: KD 있음 vs 없음
]

chi2, p_val, dof, expected = chi2_contingency(contingency_table)

# 4) 결과 출력
print("=== KD Occurrence (Binary) Chi-square Test ===")
print(f"KO KD count:   {ko_kd_count} / {len(ko_data)}")
print(f"NKO KD count:  {nko_kd_count} / {len(nko_data)}")
print(f"\nChi-square: {chi2:.4f}")
print(f"p-value:     {p_val:.6f}")
print(f"Degrees of freedom: {dof}")
print("\nExpected frequencies:")
print(expected)

# 5) 시각화 - 비율 비교
labels = ['KD Occurred', 'No KD']
ko_vals = [ko_kd_count, ko_no_kd]
nko_vals = [nko_kd_count, nko_no_kd]

plt.figure(figsize=(8,5))
bar_width = 0.35
x = np.arange(2)

plt.bar(x - bar_width/2, ko_vals,  width=bar_width, label='KO')
plt.bar(x + bar_width/2, nko_vals, width=bar_width, label='NKO')

plt.xticks(x, labels)
plt.title("KD Occurrence Rate (KO vs NKO)")
plt.legend()
plt.show()
```
